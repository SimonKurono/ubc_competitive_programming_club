Blackslex and Password

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'

def backslex_and_password(k, x):
    # Your implementation here
    usable = alphabet[:k]

    res = ''
    n = 0

    while True:
        start = len(res)

        for c in usable:
            temp = res+c
            if (checkX(temp, x)):
                res += c
                n += 1
            
        if (n == start): 
            break


    return n+1

def checkX(s, x):
    j = len(s)  - 1

    for i in range(0, j):
        if ((j-i) > 0 and (j-1) % x == 0 and s[i] == s[j]):
            return False
            
    return True
```

```python
print (backslex_and_password(2, 1))
print (backslex_and_password(3, 2))
print (backslex_and_password(1, 5))
```
